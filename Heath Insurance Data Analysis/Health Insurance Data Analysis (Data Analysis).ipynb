```python
Obesity_data.head(2)
```

<h1>Analysis of Cholesterol Values in Participants

```python
#Cholesterol distribution

#Health Cholesterol values -
#Total Cholesterol - less than 200 mg/dL
#LDL Cholesterol - less than 130 mg/dL
#HDL Cholesterol - 50 and higher mg/dL
#Triglycerides - Less than 200 mg/dL
```

```python
#Created a column as CHOLESTEROL LEVEL
#0 - Health and 1 - Unhealthy
```

```python
def cholesterol(x):
    if x['TOTAL CHOLESTEROL']>=200 or x['TRIGLYCERIDE']>=130 or x['HDL CHOLESTEROL']<50 or x['LDL CHOLESTEROL']>=200:
        return 1
    else:
        return 0
```

```python
Data['CHOLESTEROL LEVEL'] = Data.apply(cholesterol, axis=1)
```

<h2>Pivot Table to Show Percentage of Participants with High Cholesterol Values by Age Group

```python
pivot = Data.pivot_table(index='AGE',columns='CHOLESTEROL LEVEL',values='SEX',aggfunc='count',margins=True).reset_index()
```

```python
pivot.drop(14,inplace=True)
```

```python
pivot[0]= (pivot[0]/pivot['All'])*100
pivot[1]= (pivot[1]/pivot['All'])*100
```

```python
pivot[[0,1]]=pivot[[0,1]].round(decimals=1)
pivot
```

<h2>Percentage of Participants with High Cholesterol and Normal Values by Age

```python
ax=pivot[['AGE',1,0]].plot.bar(x='AGE',stacked=True,figsize=(10,5))
ax.bar_label(ax.containers[0])
ax.legend(['High Cholesterol','Normal'])
plt.ylabel('Count(%)')
plt.xlabel('Age')
plt.title('Participants with High Cholesterol by Age')
plt.show()
```

<h2>Average Cholesterol Levels Across Age Groups

```python
Data.pivot_table(index='AGE',values=['TOTAL CHOLESTEROL','TRIGLYCERIDE','HDL CHOLESTEROL','LDL CHOLESTEROL'],
                 aggfunc='mean').round(decimals=1).plot.line(figsize=(10,5))
plt.ylabel('Cholesterol Value(mg/dL)')
plt.xlabel('Age')
plt.title('Average Cholesterol Levels Across Age Groups')
plt.show()
```

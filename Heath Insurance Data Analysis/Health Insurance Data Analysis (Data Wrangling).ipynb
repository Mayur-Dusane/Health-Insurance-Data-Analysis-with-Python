<h1>Data Cleaning and Preprocessing

```python
#Renaming column names for better understanding
Clean_Health_Data = health_data.rename(columns={'IDV_ID':'ID', 'HEIGHT':'HEIGHT(CM)', 'WEIGHT':'WEIGHT(KG)','WAIST':'WAIST(CM)',
                            'SIGHT_LEFT':'EYE-SIGHT(LEFT)','SIGHT_RIGHT':'EYE-SIGHT(RIGHT)','HEAR_LEFT':'HEARING(LEFT)',
                            'HEAR_RIGHT':'HEARING(RIGHT)','BP_HIGH':'SYSTOLIC BP','BP_LWST':'DIASTOLIC BP','BLDS':'FASTING BLOOD SUGAR',
                            'TOT_CHOLE':'TOTAL CHOLESTEROL','HDL_CHOLE':'HDL CHOLESTEROL','LDL_CHOLE':'LDL CHOLESTEROL',
                            'HMG':'HEMOGLOBIN','OLIG_PROTE_CD':'URINE PROTEIN','SGOT_AST':'AST','SGPT_ALT':'ALT','GAMMA_GTP':'GAMMA GTP',
                            'SMK_STAT':'SMOKING','DRK_YN':'ALCOHOL','HCHK_CE_IN':'ORAL EXAMINATION','CRS_YN':'DENTAL CARIES','TTR_YN':'CALCULUS TEETH'})
```

```python
Clean_Health_Data.head(2)
```

<b>Dropping Unwanted Columns.<br>
Removing columns ('YEAR','AREA_CODE','ORAL EXAMINATION', 'DENTAL CARIES','CALCULUS TEETH','DATE').

```python
#Removing unwanted Columns
Clean_Health_Data.drop(['YEAR','AREA_CODE','ORAL EXAMINATION', 'DENTAL CARIES','CALCULUS TEETH','DATE'], axis=1, inplace=True)
```

```python
#Changing datatypes of columns
Clean_Health_Data[['HEARING(LEFT)','HEARING(RIGHT)','SYSTOLIC BP','DIASTOLIC BP','FASTING BLOOD SUGAR','TOTAL CHOLESTEROL',
                   'TRIGLYCERIDE','HDL CHOLESTEROL','LDL CHOLESTEROL', 'URINE PROTEIN', 'CREATININE', 'AST', 'ALT',
                   'GAMMA GTP', 'SMOKING', 'ALCOHOL']] = Clean_Health_Data[['HEARING(LEFT)','HEARING(RIGHT)','SYSTOLIC BP', 'DIASTOLIC BP', 'FASTING BLOOD SUGAR','TOTAL CHOLESTEROL', 
                   'TRIGLYCERIDE', 'HDL CHOLESTEROL', 'LDL CHOLESTEROL', 'URINE PROTEIN','CREATININE', 'AST', 'ALT', 
                   'GAMMA GTP', 'SMOKING', 'ALCOHOL']].astype(int)
```

<b>Creating 'AGE' column with the help of 'AGE_GROUP' column.</b>

```python
#Replacing AGE_GROUP column with AGE column
#AGE=AGE_GROUP * 5
#AGE will be multiple of 5 according to dataset
```

```python
Clean_Health_Data['AGE_GROUP']=Clean_Health_Data['AGE_GROUP']*5
Clean_Health_Data.rename(columns={'AGE_GROUP':'AGE'}, inplace=True)
```

```python
Clean_Health_Data.head(2)
```

<b>Replacing SEX column numerical values with categorical values.</b>

```python
#Replacing SEX column numerical values with categorical values
Clean_Health_Data['SEX'] = Clean_Health_Data['SEX'].replace({1:'M',2:'F'})
```

<b>Changing 'SMOKING' column values.</b><br>

```python
#Changing SMOKING columns values
Clean_Health_Data['SMOKING'] = Clean_Health_Data['SMOKING'].replace({1:0,2:1})
```

<b>Changing 'HEARING(LEFT)' and 'HEARING(RIGHT)' columns values to 0 (No hearing impairment) and 1 (Hearing impairment).</b>

```python
#Changing HEARING(LEFT) values to 0 (No hearing impairment) and 1 (Hearing impairment)
Clean_Health_Data['HEARING(LEFT)'] = Clean_Health_Data['HEARING(LEFT)'].replace({1:0,2:1})
```

```python
#Changing HEARING(RIGHT) values to 0 (No hearing impairment) and 1 (Hearing impairment)
Clean_Health_Data['HEARING(RIGHT)'] = Clean_Health_Data['HEARING(RIGHT)'].replace({1:0,2:1})
```

<b>Creating 'BMI' Column using 'HEIGHT(CM)' and 'WEIGHT(KG)' columns.<b/>

```python
#Creating BMI Column using HEIGHT(CM) and WEIGHt(KG) columns
Clean_Health_Data['BMI']= Clean_Health_Data['WEIGHT(KG)']/(Clean_Health_Data['HEIGHT(CM)']/100)**2
```

```python
#Rounding off BMI Column values upto one decimal
Clean_Health_Data['BMI'] = Clean_Health_Data['BMI'].round(decimals=1)
```

<b>Creating 'W/H RATIO' (Waist to height ratio) column using 'WAIST(CM)' and 'HEIGHT(CM)' columns.<b>

```python
#Creating WAIST TO HEIGHT RATIO
Clean_Health_Data['W/H RATIO'] = Clean_Health_Data['WAIST(CM)']/Clean_Health_Data['HEIGHT(CM)']
Clean_Health_Data['W/H RATIO'] = Clean_Health_Data['W/H RATIO'].round(decimals=2)
```

```python
Clean_Health_Data.head()
```

<b>Creating HEARING IMPAIRMENT column by comparing HEARING(LEFT) and HEARING(RIGHT).

```python
#Creating HEARING IMPAIRMENT column by comparing HEARING(LEFT) and HEARING(RIGHT)
Clean_Health_Data['HEARING IMPAIRMENT'] = Clean_Health_Data.apply(lambda x: 1 if (x['HEARING(LEFT)']==1 or x['HEARING(RIGHT)']==1) else 0, axis=1)
```

```python
#Rearranging columns
Final_Data = Clean_Health_Data.loc[:,['SEX', 'AGE', 'HEIGHT(CM)', 'WEIGHT(KG)', 'WAIST(CM)', 'BMI','W/H RATIO',
       'EYE-SIGHT(LEFT)', 'EYE-SIGHT(RIGHT)', 'HEARING(LEFT)',
       'HEARING(RIGHT)','HEARING IMPAIRMENT', 'SYSTOLIC BP', 'DIASTOLIC BP', 'FASTING BLOOD SUGAR',
       'TOTAL CHOLESTEROL', 'TRIGLYCERIDE', 'HDL CHOLESTEROL',
       'LDL CHOLESTEROL', 'HEMOGLOBIN', 'URINE PROTEIN', 'CREATININE', 'AST',
       'ALT', 'GAMMA GTP', 'SMOKING', 'ALCOHOL',
       ]]
```
